## Uvalde

### Scrape

```python
# Creating list to append tweet data
uvalde_all_tweets_list = []
uvalde_hash_tweets_list = []
uvalde_source_tweets_list = []
```

```python
# Using TwitterSearchScraper to scrape data and append tweets to list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('(uvalde OR "robb elementary" OR #uvalde OR #robbelementary) AND (#rip OR robb OR elementary OR mass OR school OR shooting OR massacre OR shooter ) min_faves:10 filter:has_engagement until:2022-06-07 since:2022-05-24').get_items()): #declare search criteria
    if i > 1000:  # number of tweets you want to scrape
        break
    try:
        uvalde_all_tweets_list.append([tweet.id, tweet.date, tweet.sourceLabel, tweet.hashtags, tweet.content, tweet.lang,
                                tweet.user.username, tweet.retweetCount,tweet.quoteCount, tweet.likeCount, tweet.user.followersCount, tweet.mentionedUsers])   # declare the attributes to be returned
        uvalde_hash_tweets_list.append([tweet.id, tweet.hashtags, tweet.mentionedUsers, tweet.user.followersCount, tweet.user.username]) 
        uvalde_source_tweets_list.append([tweet.sourceLabel,tweet.retweetCount, tweet.quoteCount, tweet.likeCount])

    except:
        continue 
```

```python
#Twitter Field Directory
print(dir(tweet))
```

```python
# Creating a dataframe from the tweets list above
uvalde_all_tweets_df = pd.DataFrame(uvalde_all_tweets_list, columns=['Tweet Id','DT', 'sourceLabel', 'Hashtags', 'Text', 'lang','User','RTcount','quoteCount', 'likeCount','followersCount', 'Mentioned'])
                     
uvalde_hash_tweets_df = pd.DataFrame(uvalde_hash_tweets_list, columns=['Tweet Id','Hashtags','Mentioned', 'FollowerCount','User'])

uvalde_source_tweets_df = pd.DataFrame(uvalde_source_tweets_list, columns=['sourceLabel','RTcount','quoteCount','likeCount'])

#check number of tweets pulled
uvalde_all_tweets_df.count()
```

## Token/Sentiment Analysis

### Setup

```python
nltk.download('stopwords')
nltk.download('punkt')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import word_tokenize

#create variable for tokenizer
ttokenizer = nltk.tokenize.TweetTokenizer()

nltk_stopwords = nltk.corpus.stopwords.words('english')

#Add extra stopwords to list
otherstopwords = [':','.','!',';','?',',','*','^','_','(',")","'",'-','"','&','...','"',"'"]

nltk_stopwords.extend(otherstopwords)


sid = SentimentIntensityAnalyzer()
def get_sentiment(text:str, analyser,desired_type:str='pos'):
    # Get sentiment from text
    sentiment_score = analyser.polarity_scores(text)
    return sentiment_score[desired_type]

# Get Sentiment scores
def get_sentiment_scores(df,data_column):
    df[f'{data_column} Positive Sentiment Score'] = df[data_column].astype(str).apply(lambda x: get_sentiment(x,sid,'pos'))
    df[f'{data_column} Negative Sentiment Score'] = df[data_column].astype(str).apply(lambda x: get_sentiment(x,sid,'neg'))
    df[f'{data_column} Neutral Sentiment Score'] = df[data_column].astype(str).apply(lambda x: get_sentiment(x,sid,'neu'))
    df[f'{data_column} Compound Sentiment Score'] = df[data_column].astype(str).apply(lambda x: get_sentiment(x,sid,'compound'))
    return df

#view
nltk_stopwords[:10]
```

#### Tweets by Tokenizaiton

```python
#Create datafram of text again back from source
uvalde_tok_text = pd.DataFrame(uvalde_textlist)

#Clean data by removing hyperlinks and 
#hashtags(only remove # as we are interested in seeing all the popular words including hash values)
#leave @ to see how many times someone is mentioned 
uvalde_tok_text['check'] = uvalde_tok_text[0].apply(lambda x : re.sub(r'http\S+', '', x))
uvalde_tok_text['check'] = uvalde_tok_text['check'].apply(lambda x : re.sub(r'#+', '', x))


#Use special tokenizer to keep special words/phases together (ie. gun control rather than gun and control)
uvalde_sw_tokenizer = nltk.MWETokenizer([('gun','control'),('elementary','school')])


uvalde_tokcheck =[]

#Apply tokenizer to each line of text
for line in uvalde_tok_text['check']:
    line = ttokenizer.tokenize(line)
    line = uvalde_sw_tokenizer.tokenize((line))
    uvalde_tokcheck.extend(line)



uvalde_cleantok = []

#Remove stopwords
for line in uvalde_tokcheck:
    line = line.lower()
    if not line in nltk_stopwords:
        uvalde_cleantok.append(line)
    else:
        continue

#view 
#uvalde_cleantok
```

```python
#Find frequency of words in text
uvalde_tokFD = nltk.FreqDist(uvalde_cleantok)
uvalde_top_toks = uvalde_tokFD.most_common()
# for word, freq in uvalde_top_toks:
#     print(word, freq)

uvalde_top_toksdf = pd.DataFrame(uvalde_top_toks).rename(columns={0:'Uvalde Text',1:'Uvalde Freq'}).sort_values(by=['Uvalde Freq'],ascending=False).reset_index().drop(columns=['index'])
uvalde_top_toksdf[:10]
```
